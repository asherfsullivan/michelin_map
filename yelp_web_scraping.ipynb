```python
import requests, bs4, time, csv, math
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#get cities from ca michelin dataset
ca_michelin_data_to_load = ('./Resources/ca_michelin_data.csv')

# read the data
ca_michelin_df = pd.read_csv(ca_michelin_data_to_load)

ca_cities_list = ca_michelin_df["city"].unique()
for city in ca_cities_list:
    print(city)
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
urls = ['https://www.yelp.com/search?find_desc=restaurants&find_loc=San+Francisco%2C+CA&ns=1&attrs=RestaurantsPriceRange2.2%2CRestaurantsPriceRange2.3%2CRestaurantsPriceRange2.4',
       'https://www.yelp.com/search?find_desc=restaurants&find_loc=San%20Francisco%2C%20CA&ns=1&attrs=RestaurantsPriceRange2.2%2CRestaurantsPriceRange2.3%2CRestaurantsPriceRange2.4&start=10',
       'https://www.yelp.com/search?find_desc=restaurants&find_loc=Monterey%2C+CA&attrs=RestaurantsPriceRange2.2%2CRestaurantsPriceRange2.3%2CRestaurantsPriceRange2.4',
       'https://www.yelp.com/search?find_desc=restaurants&find_loc=Sacramento%2C+CA&attrs=RestaurantsPriceRange2.2%2CRestaurantsPriceRange2.3%2CRestaurantsPriceRange2.4',
       'https://www.yelp.com/search?find_desc=restaurants&find_loc=Los+Angeles%2C+CA&attrs=RestaurantsPriceRange2.2%2CRestaurantsPriceRange2.3%2CRestaurantsPriceRange2.4',
       'https://www.yelp.com/search?find_desc=restaurants&find_loc=Costa+Mesa%2C+CA&attrs=RestaurantsPriceRange2.2%2CRestaurantsPriceRange2.3%2CRestaurantsPriceRange2.4',
       'https://www.yelp.com/search?find_desc=restaurants&find_loc=San+Diego%2C+CA&attrs=RestaurantsPriceRange2.2%2CRestaurantsPriceRange2.3%2CRestaurantsPriceRange2.4']
```

```python
#get Yelp data 

name_list = []
street_list = []
cuisine_list = []
price_list = []
rating_list = []
review_count_list = []

for url in urls:
    browser.visit(url)

    for i in range(4,12):
        link = browser.find_by_css('a[class="css-1m051bw"]')[i]
        link.click()
        #restaurant name
        restaurant_name = browser.find_by_css('h1[class="css-12dgwvn"]').text
        name_list.append(restaurant_name)
        #street
        a = browser.find_by_css('p[class=" css-r9996t"]')
        street = a.find_by_css('span[class=" raw__09f24__T4Ezm"]').text
        street_list.append(street)
        #cuisine
        cuisine = browser.find_by_css('a[class="css-1m051bw"]').text
        cuisine_list.append(cuisine)
        #price
        price = browser.find_by_css('span[class=" css-1ir4e44"]').text
        price_list.append(price)
        #rating
        element = browser.find_by_xpath(".//div[contains(@class,'i-stars')]")
        rating = element['aria-label']
        rating_list.append(rating)
        #review count
        sidebar = browser.find_by_css('div[class=" arrange-unit__09f24__rqHTg arrange-unit-fill__09f24__CUubG border-color--default__09f24__NPAKY nowrap__09f24__lBkC2"]')
        review_count = sidebar.first.find_by_css('span[class=" css-1fdy0l5"]').text
        review_count_list.append(review_count)

        browser.back()
```

```python
browser.quit()
```

```python
yelp_df = pd.DataFrame()

yelp_df["name"] = name_list
yelp_df["street"] = street_list
yelp_df["cuisine"] = cuisine_list
yelp_df["review count"] = review_count_list
yelp_df["rating"] = rating_list
yelp_df["price rating"] = price_list

yelp_df
```

```python
address_list = []

for url in urls:
    browser.visit(url)

    for i in range(4,12):
        link = browser.find_by_css('a[class="css-1m051bw"]')[i]
        link.click()
        #city 
        c = browser.find_by_css('img[alt="Map"]')
        c.click()
        address = browser.find_by_tag('address').text
        address_list.append(address)
        browser.visit(url)
```
