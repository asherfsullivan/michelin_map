```python
# import libraries
import pandas as pd
from pathlib import Path 
import json
```

```python
# load data from json files
business_data_to_load = ('./Resources/yelp_academic_dataset_business.json')

# read data into dfs
business_df = pd.read_json(business_data_to_load, lines=True)
```

```python
#show df
business_df.head()
```

```python
# get restaurant names as a list

names = business_df.loc[:,"name"].to_list()
```

```python
# load michelin data
michelin_data_to_load = ('./Resources/michelin_data.csv')

# read the data
michelin_df = pd.read_csv(michelin_data_to_load)
```

```python
michelin_df.head()
```

```python
#find all michelin restaurants that have a yelp review in our data file
both_df = michelin_df.where(michelin_df['name'].isin(names))
```

```python
#drop NaN values
both_df = both_df.dropna()
both_df = both_df[(both_df.region != 'Austria')
                  & (both_df.region != 'Finland')
                  & (both_df.region != 'Norway')
                  & (both_df.region != 'Hungary')]
both_df
```

```python
# make a list of names on both_df and make a df with only those names from yelp data and merge the 2 dfs
both_names = both_df.loc[:,"name"].to_list()

yelp_names_df = business_df.where(business_df['name'].isin(both_names)).dropna()

yelp_names_df
```

```python
ca_michelin_df = michelin_df.where(michelin_df['region'] == 'California').dropna()
                        
ca_michelin_df = ca_michelin_df.reset_index(drop=True)

ca_michelin_df
```
